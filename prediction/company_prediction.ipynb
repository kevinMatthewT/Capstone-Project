```python
from prophet import Prophet
import matplotlib.pyplot as plt

# Function to forecast using Prophet for each column
def forecast_with_prophet(df_to_predict, column, forecast_days):
    # Prepare the data for Prophet
    forecast_data = df_to_predict[[column]].reset_index()
    forecast_data.columns = ['ds', 'y']  # Rename columns to fit Prophet's requirement
    
    # Remove any timezone information from the 'ds' column
    forecast_data['ds'] = forecast_data['ds'].dt.tz_localize(None)
    
    # Drop rows with NaN values in 'ds' or 'y'
    forecast_data = forecast_data.dropna(subset=['ds', 'y'])
    
    # Initialize Prophet model
    model = Prophet()

    # Fit the model
    model.fit(forecast_data)

    # Make future dataframe for predictions (Do NOT pass periods twice)
    future = model.make_future_dataframe(forecast_data, periods=forecast_days)

    # Predict
    forecast = model.predict(future)
    
    # Plot the forecast with historical data and predicted data
    fig = model.plot(forecast)
    plt.title(f"Forecast for {column}")
    plt.xlabel('Date')
    plt.ylabel(column)
    
    # Highlight historical data and forecasted data
    plt.axvline(x=forecast_data['ds'].iloc[-1], color='red', linestyle='--', label='Forecast Start')

    # Show the plot
    plt.legend()
    plt.show()

    # Return forecasted data for use if needed
    return forecast

# Forecasting for each column
forecast_days = 90  # Forecasting the next 90 days
for column in df_to_predict.columns:
    forecast_with_prophet(df_to_predict, column, forecast_days)
```
