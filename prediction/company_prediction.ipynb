```python
import requests
import json
```

```python
unfiltered_endpoint="/api/get/investment/Company/filter/"
company_search="Foodies Restaurant"

url=f"http://localhost:8080{unfiltered_endpoint}{company_search}"
```

```python
body=requests.get(url).json()
```

```python
import pandas as pd
df=pd.DataFrame(body)
```

```python
df
```

```python
df.drop(columns=["_id","__v"],inplace=True, errors='ignore')
df
```

```python
df["Date_Of_Ownership"]=pd.to_datetime(df["Date_Of_Ownership"])
df["Date_Of_Ownership"]=df["Date_Of_Ownership"].dt.date

df
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```
